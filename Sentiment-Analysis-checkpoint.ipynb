## ETAPE 3: Text Preprocessing

*** création d'une fonction qui regroupera les étapes clés du traitement du texte. Cette fonction préparera les deux jeux de données de manière cohérente, rendant le texte prêt pour les modèles de machine learning et deep learning.***

```python
# Télécharger les stopwords si nécessaire
nltk.download('stopwords')

# Fonction de prétraitement
def preprocess_text(df, text_column):
    STOPWORDS = set(stopwords.words('english'))
    
    # Convertir la colonne en chaîne (si nécessaire)
    df[text_column] = df[text_column].astype(str)
    
    # Étendre les contractions
    df[text_column] = df[text_column].apply(contractions.fix)
    
    # Supprimer les chiffres
    df[text_column] = df[text_column].apply(lambda x: ''.join([char for char in x if not char.isdigit()]))
    
    # Convertir en minuscule
    df[text_column] = df[text_column].str.lower()
    
    # Supprimer les balises HTML
    df[text_column] = df[text_column].apply(lambda x: re.sub(r'<.*?>', ' ', x))
    
    # Supprimer les ponctuations
    df[text_column] = df[text_column].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))
    
    # Supprimer les caractères spéciaux
    df[text_column] = df[text_column].apply(lambda x: re.sub(r'[^a-zA-Z0-9\s]', '', x))
    
    # Supprimer les stopwords
    df[text_column] = df[text_column].apply(lambda x: " ".join([word for word in x.split() if word not in STOPWORDS]))
    
    # Supprimer les espaces multiples
    df[text_column] = df[text_column].apply(lambda x: re.sub(r'\s+', ' ', x).strip())
    
    return df
```

## 1.  Analyse des Sentiments avec TextBlob

Dans cette section, nous allons utiliser **TextBlob** pour analyser chaque phrase de notre ensemble de données. TextBlob est un outil puissant de traitement du langage naturel qui calcule la **polarité** des phrases. La polarité se décompose en deux éléments principaux : **polarité** et **subjectivité**.

### 1. Polarité
La **polarité** est une valeur numérique qui identifie si une phrase exprime un sentiment **positif** ou **négatif**. Elle varie entre **-1 et 1**, où :

- **1** indique un **sentiment très positif** ou une affirmation basée sur des **faits** plutôt que des émotions.  
- **-1** reflète un **sentiment très négatif** ou une affirmation sans fondement factuel.

En résumé :  
- Plus la polarité est proche de **1**, plus le sentiment est positif.  
- Plus elle est proche de **-1**, plus le sentiment est négatif.

### 2. Subjectivité
La **subjectivité** mesure dans quelle mesure une phrase reflète des **opinions personnelles** ou des **émotions**, par opposition à des **faits objectifs**. Ses valeurs oscillent entre **0 et 1** :

- **1** indique une forte subjectivité, c’est-à-dire une phrase fondée sur des **opinions** ou des **émotions**.  
- **0** suggère une forte objectivité, indiquant une phrase basée sur des **faits**.

### Résumé
La polarité évalue le **ton émotionnel** (positif ou négatif) d’une phrase, tandis que la subjectivité détermine si l’énoncé repose sur des **opinions** ou des **faits**. Ces deux métriques combinées nous permettent d’obtenir une analyse plus riche des sentiments véhiculés dans les textes.

```python
# Tokenisation, Pos_tagging et Lemmatisation

nlp = spacy.load('en_core_web_sm') # Charger le modèle spaCy

# Définir une fonction pour la Tokenisation, Pos_tagging et Lemmatisation
def lemmatize_text(text):
    doc = nlp(text)  # Traiter le texte avec spaCy
    # Remplacer les tokens par leurs lemmes
    return ' '.join([token.lemma_ if token.lemma_ != '-PRON-' else token.text for token in doc])
```

Rappelons que l'analyse des sentiments avec Texblob vise à différencier les messages factuels des messages émotionnels.  

- Les **messages factuels** sont généralement représentés par des **valeurs positives**.  
  - Ils reflètent des faits réels, souvent soutenus par des preuves concrètes.  
  - Exemple : Une personne anxieuse dit qu'elle est anxieuse parce qu'elle s'appuie sur des événements ou des circonstances tangibles qui justifient son ressenti.  

- En revanche, les **messages émotionnels** sont souvent représentés par des **valeurs négatives**.  
  - Ces messages expriment un ressenti dicté principalement par les émotions, sans preuves tangibles.  
  - Exemple : Une personne peut déclarer être en dépression non pas parce qu'elle l'est réellement, mais par empathie ou émotion, s'identifiant à la souffrance d'autrui.  

Cependant, cette analyse ne garantit pas une vérité absolue. Elle offre plutôt une méthode pour évaluer, en moyenne, la véracité ou l'origine des messages, tout en tenant compte de la nuance entre émotion et réalité.

```python
# Tokenization train_df

texts = df_DeepLearning['statement'].values  # Les textes des patients
labels = df_DeepLearning['status'].values  # Les états des patients

# Conversion des étiquettes 
label_mapping = {
    'Normal': 0,
    'Depression': 1,
    'Suicidal': 2,
    'Anxiety': 3,
    'Bipolar': 4,
    'Stress': 5,
    'Personality disorder': 6
}
labels = df_DeepLearning['status'].map(label_mapping).values


# Initialisation des parametres

maxlen = 100 # Utiliser les 100 premiers mots dans un commentaire
training_samples = 200 # 200 entrées pour entrainer le model
validation_samples = 10000  # 10000 entrées pour l'évaluation
max_words = 10000 # Par exemple, prendre les 10 000 mots les plus fréquents

tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)

word_index = tokenizer.word_index
print('Found %s unique tokens.' % len(word_index))

# Padding
data = pad_sequences(sequences, maxlen=maxlen)
labels = np.asarray(labels)
print('shape of data tensor', data.shape)
print('shape of label tensor:', labels.shape)

# Rearranger aléatoirement les données
indices = np.arange(data.shape[0])
np.random.shuffle(indices)
data = data[indices]
labels = labels[indices]

# Séparation des données en ensemble d'entrainement et ensemble de validation
X_train = data[:training_samples] # 200 premier echantillons pour l'entrainement
y_train = labels[:training_samples] # 200 premier labels pour l'entrainement
X_val = data[training_samples: training_samples + validation_samples] # 10000 premier échantillons pour l'evaluation
y_val = labels[training_samples: training_samples + validation_samples] # 10000 labels pour l'évaluation
```

```python
# Téléchargement  du  model pré-entrainé  Glove

# Chemin vers les embeddings GloVe
glove_file = "C:/Users/claud/Documents/Machine learning MIT/Sentiment_Analysis/glove.6B/glove.6B.100d.txt"

# Chargement des embeddings GloVe
embeddings_index = {}
f = open(glove_file, encoding="utf8")  
for line in f:
    values = line.split()
    word = values[0]
    coefs = np.asarray(values[1:], dtype='float32')
    embeddings_index[word] = coefs
f.close()

print('Found %s word vectors.' % len(embeddings_index))
```

```python
# Tester si le charement a été bien éffectué

embeddings_index['good']
```

```python
# initialisation de la  matrice d'embedding pour le modèle 

embedding_dim = 100
embedding_matrix = np.zeros((max_words, embedding_dim))
for word, i in word_index.items():
    embedding_vector = embeddings_index.get(word)
    if i < max_words:
        if embedding_vector is not None:
            embedding_matrix[i] = embedding_vector
```

```python
# Creation du Model de Deep Learning
from tensorflow.keras.layers import Input

model = Sequential()
model.add(Input(shape=(maxlen,))) 
model.add(Embedding(max_words, embedding_dim, input_length = maxlen))
model.add(Flatten())
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(32, activation='sigmoid'))
model.add(Dense(7, activation='softmax'))
model.summary()
```

```python
model.layers[0].set_weights([embedding_matrix]) # Initialisation des poids de la couche d'embedding :
model.layers[0].trainable = False # Empêcher l’entraînement de la couche d’embedding :
```

```python
#Compiler le modele

model.compile(
    loss='categorical_crossentropy',  # 'categorical_crossentropy' pour une classification multi-classes
    optimizer = 'Adam', 
    metrics=['accuracy']  # Mesure la précision pendant l'entraînement
)
```

##### Pourquoi stratifier ?
La stratification est essentielle lorsque vous travaillez avec des classes déséquilibrées (par exemple, si certaines catégories comme 
***Dépression*** ou ***Stress*** apparaissent beaucoup plus fréquemment que d'autres). Cela garantit que chaque ensemble de données (entraînement et test) contient une distribution proportionnelle des classes.

```python
# Fonction d'évaluation du modéle

def metrics_score(actual, predicted, classes):
    # Générer le rapport de classification sous forme de DataFrame
    report = classification_report(actual, predicted, target_names=classes, output_dict=True)
    report_df = pd.DataFrame(report).transpose()
    
    return report_df

```

```python
# Fonction pour calculer les poids

def compute_weight(df, target):
    values = target.unique()  # Obtenir les valeurs uniques dans target (les classes)
    
    for value in values:
        n_classes = target.nunique()  # Nombre de classes uniques
        n_samples = df.shape[0]  # Nombre total d'échantillons
        class_samples = target[target == value].shape[0]  # Nombre d'échantillons pour la classe 'value'

        # Calcul du poids pour chaque classe
        class_weight = round(n_samples / (n_classes * class_samples), 2)
        
        print(f"Poids pour la classe '{value}': {class_weight}")
        
compute_weight(df, df['status'])
```

Nous avons constaté que la distribution des classes est déséquilibrée dans notre jeu de données. Dès le départ, afin d'améliorer les performances de notre modèle, nous avons décidé d'attribuer des poids spécifiques à chaque classe.

Ces poids sont calculés à l'aide de la formule suivante :

\[
w_j = \frac{n_{\text{samples}}}{n_{\text{classes}} \times n_{\text{samples}_j}}
\]

### Où :
- \(w_j\) : Poids attribué à la classe \(j\).
- \(n_{\text{samples}}\) : Nombre total d'échantillons (lignes) dans le jeu de données.
- \(n_{\text{classes}}\) : Nombre total de classes uniques dans la variable cible.
- \(n_{\text{samples}_j}\) : Nombre d'échantillons appartenant à la classe \(j\).

> **Remarque** : Ce résultat peut également être obtenu automatiquement en définissant le paramètre 
`class_weight='balanced'` dans de nombreux algorithmes d'apprentissage supervisé.

```python
# Prétraitement du texte
class TextPreprocessor:
    def __init__(self):
        self.stopwords = set(stopwords.words('english'))
    
    def preprocess(self, text):
        # Assurez-vous que le texte est une chaîne de caractères avant d'appliquer les transformations
        if isinstance(text, str):
            # Étendre les contractions
            text = contractions.fix(text)
            
            # Supprimer les chiffres
            text = ''.join([char for char in text if not char.isdigit()])
            
            # Convertir en minuscule
            text = text.lower()
            
            # Supprimer les balises HTML
            text = re.sub(r'<.*?>', ' ', text)
            
            # Supprimer les ponctuations
            text = text.translate(str.maketrans('', '', string.punctuation))
            
            # Supprimer les caractères spéciaux
            text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
            
            # Supprimer les stopwords
            text = " ".join([word for word in text.split() if word not in self.stopwords])
            
            # Supprimer les espaces multiples
            text = re.sub(r'\s+', ' ', text).strip()
        
        return text

# Transformer pour gérer les textes invalides
class InvalidTextHandler(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self  # Pas d'opération de "fit" nécessaire
    
    def transform(self, X):
        # Assurez-vous que chaque texte dans X est une chaîne avant de le traiter
        return [
            "texte_non_valide" if not isinstance(text, str) or text.strip() == "" or re.match(r'^[\W_]+$', text) else text
            for text in X
        ]

# Définir votre RandomForest Classifier (ou tout autre modèle que vous souhaitez utiliser)
rf_estimator = RandomForestClassifier(n_estimators=100, random_state=42)

# Pipeline complet
pipeline = Pipeline([
    # Étape de prétraitement du texte
    ('text_preprocessing', FunctionTransformer(
        TextPreprocessorFunction(), validate=False)),  # Prétraitement du texte
    
    # Gestion des textes invalides
    ('handle_invalid', InvalidTextHandler()),  # Gestion des textes invalides
    
    # Vectorisation avec TF-IDF
    ('vectorizer', TfidfVectorizer(stop_words='english', max_features=5000, ngram_range=(1, 2))),  # Vectorisation TF-IDF
    
    # Classificateur
    ('classifier', rf_estimator)  # Classification avec RandomForest
])


# Entraîner le pipeline
pipeline.fit(X_train, y_train)
```
